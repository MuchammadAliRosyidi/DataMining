```python
import numpy as np
import pandas as pd
from sklearn import tree
```

```python
irisDataset = pd.read_csv('Iris.csv', delimiter=';', header=0)
```

```python
irisDataset.head()
```

```python
print(irisDataset.columns)
```

```python
irisDataset[['Id','SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species']] = irisDataset['Id,SepalLengthCm,SepalWidthCm,PetalLengthCm,PetalWidthCm,Species'].str.split(',', expand=True)
```

```python
irisDataset["Species"] = pd.factorize(irisDataset["Species"])[0]
```

```python
irisDataset.head()
```

```python
print(irisDataset)
```

```python
irisDataset = irisDataset[['Id','SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species']].to_numpy()
```

```python
print(irisDataset)
```

```python
dataTraining = np.concatenate((irisDataset[0:40, :], irisDataset[50:90, :]),
                              axis=0)
dataTesting = np.concatenate((irisDataset[40:50, :], irisDataset[90:100, :]),
                             axis=0)
```

```python
print(dataTraining)
len(dataTraining)
```

```python
print(dataTesting)
len(dataTesting)
```

```python
inputTraining = dataTraining[:, 0:5].astype(float) # Now, dataTraining contains only numerical data and convert to float
inputTesting = dataTesting[:, 0:5].astype(float)
labelTraining = dataTraining[:, 5].astype(int)
labelTesting = dataTesting[:, 5].astype(int)
print(labelTraining)
len(labelTraining)
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model = model.fit(inputTraining, labelTraining)
```

```python
hasilPrediksi = model.predict(inputTesting)
print("Label Sebenarnya : ", labelTesting)
print("Hasil Prediksi : ", hasilPrediksi)
```

```python
prediksiBenar = (hasilPrediksi == labelTesting).sum()
prediksiSalah = (hasilPrediksi != labelTesting).sum()
print("Prediksi Benar :", prediksiBenar, "data")
print("Prediksi Salah :", prediksiSalah, "data")
print("Akurasi :", prediksiBenar/(prediksiBenar+prediksiSalah) * 100, "%")
```
